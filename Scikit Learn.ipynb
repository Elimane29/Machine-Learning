#  1) Régression linéaire 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
m=100 # On crée 100 échantillons
X = np.linspace(0,10, m).reshape(m,1)
y= X + np.random.randn(m,1)
```

```python
plt.scatter(X, y )
```

On voit que pour ce problème on a besoin d'un modèle de régression linéaire

```python
from sklearn.linear_model import LinearRegression
```

Et là on recopie les 4 lignes de code qu'on a vu

```python
model= LinearRegression()
model.fit(X,y)
model.score(X,y)
```

En fait là c'est le coefficient de détermination qui est calculé


C'est à dire le Rcarré de la méthode des moindres carrés

Il y a d'autres techniques d'évaluation sur scikit Learn tu peux les voir sur le site

```python
model.predict(X)
```

```python
predictions = model.predict(X)

plt.scatter(X,y)
plt.plot(X, predictions, c='r')
```

Dans ce cas-ci le linear regression était parfait mais dans un autre non

```python
#Pour un autre cas où ce n'est pas linéaire 
np.random.seed(0)
m=100 # On crée 100 échantillons
X = np.linspace(0,10, m).reshape(m,1)
y= X**2 + np.random.randn(m,1)
```

```python
plt.scatter(X, y )
```

```python
from sklearn.svm import SVR
```

```python
model= SVR(C=100)
model.fit(X,y)
model.score(X,y)

predictions = model.predict(X)

plt.scatter(X, y)
plt.plot(X, predictions, c='r' , lw=3)
```

Et là on est bon
